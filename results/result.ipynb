```python
def customStudentDecoder(objDict):
    return namedtuple('X', objDict.keys())(*objDict.values())

def plot_array(plot_name: str, array: set, best_epoch: int):
    plt.plot(array[:best_epoch])
    plt.ylabel(plot_name)
    plt.show()
```
